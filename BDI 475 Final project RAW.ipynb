blue First Blood by Level

```python
kd_blue = df['blueKills'] / df['blueDeaths']
df['kd_blue'] = kd_blue
df

kd_red = df['redKills'] / df['redDeaths']
df['kd_red'] = kd_red
df
```

```python
fig = px.bar(df, x='blueAvgLevel',y='blueFirstBlood',title='Blue First blood by Level')
fig.show()
```

Red First Blood By level

```python
fig = px.bar(df, x='redAvgLevel',y='redFirstBlood',title='Red First blood by Level')
fig.show()
```

```python
blue_gpm_cs = df['blueGoldPerMin'] / df['blueCSPerMin']
df['blue_gpm_cs'] = blue_gpm_cs
avg_bluegpmcs = df['blue_gpm_cs'].mean()
df['avg_bluegpmcs'] = avg_bluegpmcs
df
```

HEATMAP of BLUE Kill/Death Ration by AVG level

```python
fig = px.density_heatmap(df, x='blueAvgLevel',y='kd_blue',title='Blue K/D by AVG Level')
fig.show()
```

HEATMAP of RED Kill/Death Ratio by AVG LEVEL

```python
fig = px.density_heatmap(df, x='redAvgLevel',y='kd_red',title='Blue K/D by AVG Level')
fig.show()
```

Blue Gold Diff by avg level

```python
fig = px.area(df, x='blueAvgLevel',y='blueGoldDiff',title='Blue Gold Diff by AVG Level')
fig.show()
```

Red Gold Diff by AVG Level

```python
fig = px.area(df, x='redAvgLevel',y='redGoldDiff',title='Red Gold Diff by AVG Level')
fig.show()
```
